```python
#al observar el valor minimo como -1 en la edad se calcula el promedio sin ese valor ya que en edad no es dato real
edades_validas=datos[datos["Edad (años)"]>=0]
promedio = edades_validas["Edad (años)"].mean()
print(promedio)
```

```python
#al identificar que no se afecto casi el promedio los valores en -1 se cuentan y se observa que son pocos datos(2702), por lo mismo no afecta casi al total de datos que son mas de 1 millon
datos["Edad (años)"].value_counts().get(-1)
```

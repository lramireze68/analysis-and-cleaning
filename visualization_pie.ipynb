```python
# Se realizan los  graficos circular o de torta, teniendo en cuenta el total de las 12 columnas contenidas en el archivo final de cleaning
# Se realizan importacion de librerias
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import pandas as pd
import random
```

```python
data_cleaning = pd.read_csv("C:/Users/LILI/Desktop/GRAFICOS/Colombianos_registrados_en_el_exterior.csv", header=0)
```

```python
data_cleaning.shape
```

```python
# se realiza la grafica circular identificando la distribucion por genero e identificando que el 54.9% del genero femenino emigran a otros paise y el 45.0%son del genero masculino
distribucion_genero = data_cleaning["Género"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(distribucion_genero, labels=distribucion_genero.index,autopct="%1.1f%%", colors=["lightblue", "lightcoral", "lightgreen"])
plt.title("Distribución de Género - Colombianos en el exterior")
plt.legend()
plt.show()


```

```python
#Se realiza la distribucion por estado civil identificando que el 49.1% son solteros
distribucion_estado_civil = data_cleaning["Estado_civil"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(distribucion_estado_civil, labels=distribucion_estado_civil.index,autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Distribución estado civil - Colombianos en el exterior")
plt.legend(loc="upper right")
plt.show()
```

```python
#Por la gran cantidad de paises que se registran, se determina el top 10 de los paises mas apetecidos por los colombianos para emigrar, siendo los EE.UU el pais con mayor frecuencia de colombianos
top_10_pais = data_cleaning["País"].value_counts().nlargest(10)
plt.figure(figsize=(6, 6))
plt.title("Distribución por pais  - Colombianos en el exterior")
plt.pie(top_10_pais, labels=top_10_pais.index,autopct="%1.1f%%",startangle=180, colors=plt.cm.Paired_r.colors)
plt.legend(loc="upper right")
plt.show()
```

```python
# se realiza la distribucion por ciudad
# Se determina que Madrid y Maracaibo son las ciudades que mas registros tienen 
top_10_ciudad = data_cleaning["Ciudad_de_Residencia"].value_counts().nlargest(10)
plt.figure(figsize=(8, 8))
plt.title("Distribución por ciudad  - Colombianos en el exterior")
plt.pie(top_10_ciudad, labels=top_10_ciudad.index,autopct="%1.1f%%",startangle=180, colors=plt.cm.Paired_r.colors)
plt.show()
```

```python
#Por la gran cantidad de paises que se registran, se determina el top 10 de la ciudad de nacimiento de los colombianos 
# Se determina que el 29.8% son Bogotanos, el 20.3% corresponden a valle y el restante a otras ciudades
top_10_ciudad_nacimiento = data_cleaning["Ciudad_de_Nacimiento"].value_counts().nlargest(10)
plt.figure(figsize=(8, 8))
plt.title("Distribución por ciudad de nacimiento - Colombianos en el exterior")
plt.pie(top_10_ciudad_nacimiento, labels=top_10_ciudad_nacimiento.index,autopct="%1.1f%%",startangle=180, colors=plt.cm.Paired_r.colors)
plt.show()
```

```python
#Se realiza la distribucion por grupo de edad y se determina lo siguiente
# el 64.5 % se encuentran en la etapa de adulto 
# el 19.3 % se encuentran en la etapa de adulto mayor
# el 12.4% se encuentra en la etapa Joven
distribucion_edad = data_cleaning["Grupo_edad"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(distribucion_edad, labels=distribucion_edad.index,autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Distribución grupo edad - Colombianos en el exterior")
plt.legend()
plt.show()
```

```python
# Se realiza la distribucion por el nivel academico, identificando que el 44.0% de los colombianos no registran un nivel academico y un 24.0 % registra un nivel academico de bachillerato
distribucion_nivel_academico = data_cleaning["Nivel_Académico"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(distribucion_nivel_academico, labels=distribucion_nivel_academico.index,autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Distribución nivel academico - Colombianos en el exterior")
#se omite la leyenda debido a que ocupa parte importante del grafico plt.legend()
plt.show()
```

```python
#Se realiza la distribucion por etnia donde podemos identicar que los datos mas relevantes son:
#  El 51.4% no relaciona como "otro" en tipo de etnia
# El 43.7 % indica no tener ningun tipo de etnia 
distribucion_etnia = data_cleaning["Etnia_de_la_persona"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(distribucion_etnia, labels=distribucion_etnia.index,autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Distribución por etnia - Colombianos en el exterior")
plt.show()
```

```python
# Se realiza la distriucion por el rango de edades y se identifica que los rangos mas releantes de edades se encuentran entre:
# El 36.2 % de la poblacion se enuentra en un rango de edad entre 31 a 45 años
# El 25.6 % de la poblacion, se encuentra en un rango de edad de 46 a 60 años
# El 16.2% de la poblacion se encuentra en un rango de edad de 19 a 30 años
# El 14.0%  de la poblacion se encuentra en un rango de edad del 61 a 75
bins = [0, 18, 30, 45, 60, 75, 100]
labels = ["0-18", "19-30", "31-45", "46-60", "61-75", "76-100"]
 
data_cleaning["Rango_Edad"] = pd.cut(data_cleaning["Edad"], bins=bins, labels=labels)
distribucion_Edad = data_cleaning["Rango_Edad"].value_counts()
distribucion_Edad = distribucion_Edad.reindex(labels)
plt.figure(figsize=(6, 8))
plt.pie(distribucion_Edad, labels=distribucion_Edad.index,autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Distribución rango edades - Colombianos en el exterior")
plt.legend()
plt.show()
```

```python
# Se realiza la distriucion por el rango de estatura y se determina que:
# El rango de estatura que tienen los colombianos que emigran se encuentra entre 161 a 170 cm con un rango porcentual del 38.1%
# Seguido del rango entre 151 a 160 cm con un porcentaje del 30.1%
# Seguido del rango entre 10 a 180 cm, con un porcentaje equivalente al 21.4 % 

bins = [0, 100, 150, 160, 170, 180, 200]
labels = ["0-100", "101-150", "151-160", "161-170", "171-180", "181-200"]
 
data_cleaning["Rango_Estatura"] = pd.cut(data_cleaning["Estatura"], bins=bins, labels=labels)
distribucion_Estatura = data_cleaning["Rango_Estatura"].value_counts()
distribucion_Estatura = distribucion_Estatura.reindex(labels)
plt.figure(figsize=(6, 8))
plt.pie(distribucion_Estatura, labels=distribucion_Estatura.index,autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Distribución por estatura- Colombianos en el exterior")
plt.legend()
plt.show()
```

```python
#Se realiza la distribucion por area de conocimiento y se procede a determinar que:
# El 51.4% de los colombianos no cuenta con un area de conocimiento 
# eL 15.1% no indica o no registra areas de conocimiento
# El restante posee un area de conocimiento minima 
top_10_area_de_conocimiento = data_cleaning["Área_Conocimiento"].value_counts().nlargest(10)
plt.figure(figsize=(8, 8))
plt.title("Distribución por area de conocimiento  - Colombianos en el exterior")
plt.pie(top_10_area_de_conocimiento, labels=top_10_area_de_conocimiento.index,autopct="%1.1f%%",startangle=180, colors=plt.cm.Paired_r.colors)
plt.show()
```

```python
# Se realiza la distribucion por sub_area de conocimineto y se determina que los porcentaje mas relevantes son:
# El 61.3%  registra como ninguna en el subarea de conocimiento
# El 17.5% no indica o no responde 
# El 8.4% tiene un area de especializacion en administracion y el restante en otras areas 
top_10_subarea_de_conocimiento = data_cleaning["Sub_Area_Conocimiento"].value_counts().nlargest(10)
plt.figure(figsize=(8, 8))
plt.title("Distribución por area de conocimiento  - Colombianos en el exterior")
plt.pie(top_10_subarea_de_conocimiento, labels=top_10_subarea_de_conocimiento.index,autopct="%1.1f%%",startangle=180, colors=plt.cm.Paired_r.colors)
plt.show()
```

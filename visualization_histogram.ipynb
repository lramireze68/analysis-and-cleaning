```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import pandas as pd
```

```python
data_cleaning = pd.read_csv("C:/Users/WALTER/Desktop/posgrade/Colombianos_registrados_en_el_exterior.csv", header=0)
```

```python
data_cleaning.shape
```

```python
#es importante tener presente que para la realizacion de histogramas son mas efectivos los datos numericos en la columna, 
#por ende se realizan partiendo principalmente de las dos columnas en cleaning (edad y estatura)
plt.figure(figsize=(8, 6))
plt.hist(data_cleaning["Edad"], bins=10, color="skyblue", edgecolor="black")
#Se crea  histograma con agrupación por edades - visualization_histogram_sbs
# Etiquetas y título
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edades")

plt.show()
```

```python
plt.figure(figsize=(8, 6)) #Se crea  histograma con agrupación por estatura
data_filtrada = data_cleaning.loc[(data_cleaning["Estatura"] > 0) & (data_cleaning["Estatura"] < 220)]


plt.hist(data_filtrada["Estatura"], bins=10, color="skyblue", edgecolor="black")
#Se crea  histograma con agrupación por estatura - visualization_histogram_sbs
# Etiquetas y título
plt.xlabel("Estatura")
plt.ylabel("Frecuencia")
plt.title("Distribución de Estatura")

plt.show()
```

```python


# Se crea  histograma con agrupación de edades por género
plt.figure(figsize=(8, 6))
for genero in data_cleaning["Género"].unique(): 
    plt.hist(data_cleaning[data_cleaning["Género"] == genero]["Edad"], bins=10, alpha=0.6, label=genero)

# Etiquetas y título
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edades por Género")
plt.legend(title="Género")

plt.show()
```

```python
# Se crea  histograma con agrupación por grupo edad
plt.figure(figsize=(8, 6))
for Grupo_edad in data_cleaning["Grupo_edad"].unique():  # Iterar sobre cada categoría de género
    plt.hist(data_cleaning[data_cleaning["Grupo_edad"] == Grupo_edad]["Edad"], bins=10, alpha=0.6, label=Grupo_edad)

# Etiquetas y título
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edades por grupo edad")
plt.legend(title="Grupo_edad")

plt.show()
#gracias a este histograma podemos interpretar que las categorias para quien realizo es estudio se clasifica por edades de la siguiente forma
#adulto mayor se trata de mayores de 60, adulto entre 28 y 60, adulto joven entre 18 y 28 , adoscentes entre 12 y 18
```

```python
# Se crea  histograma con agrupación de edades por estado civil
plt.figure(figsize=(8, 6))
for Estado_civil in data_cleaning["Estado_civil"].unique():  
    plt.hist(data_cleaning[data_cleaning["Estado_civil"] == Estado_civil]["Edad"], bins=10, alpha=0.6, label=Estado_civil)

# Etiquetas y título
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edades por Estado civil")
plt.legend(title="Estado_civil")

plt.show()
```

```python
#se realiza histograma distribucion por edad visualization histogram aio
valores_edad=data_cleaning.Edad.unique()

plt.xticks(valores_edad)

formatter = FuncFormatter(lambda x, pos: f'{int(x):,}'.replace(',', '.'))

plt.gca().yaxis.set_major_formatter(formatter)

plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edades")

n, bins, patches = plt.hist(data_cleaning.Edad, color="green")

for i in range(len(n)):
    plt.text(bins[i] + 0.5, n[i] + 0.1, f'{int(n[i]):,}'.replace(',', '.'), ha='left', va='bottom', fontsize=7)

plt.show()
```

```python
#se realiza histograma distribucion por edad visualization histogram aio, se filtra por genero masculino
valores_edad=data_cleaning.Edad.unique()

plt.xticks(valores_edad)

formatter = FuncFormatter(lambda x, pos: f'{int(x):,}'.replace(',', '.'))

data_cleaning = data_cleaning[data_cleaning.Género.isin(["MASCULINO"])]

plt.gca().yaxis.set_major_formatter(formatter)

plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edades Masculino")

n, bins, patches = plt.hist(data_cleaning.Edad, color="blue")

for i in range(len(n)):
    plt.text(bins[i] + 0.5, n[i] + 0.1, f'{int(n[i]):,}'.replace(',', '.'), ha='left', va='bottom', fontsize=7)

plt.show()
```
